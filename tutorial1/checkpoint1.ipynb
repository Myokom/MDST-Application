Find the maximum outcome score for every state. 

Refer to the `Grouping and Aggregating` section in Tutorial 0 if you are stuck.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
# I import the train_test_split from Sklearn to split the datasets into subsets from training and testing
from sklearn.model_selection import train_test_split
```

```python
# I assign the the three columns to the predictor variable X and drop columns with missing values.
X = df[["SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT", "ENROLL_8", "AVG_READING_8_SCORE"]].dropna()

# I assign the AVG_MATH_8_SCORE of the same rows in the X series to Y.
y = df.loc[X.index]['AVG_MATH_8_SCORE']

# I fill the missing values in the y series with the median of the non-missing values
y.fillna(y.median(), inplace=True)
```

```python
# I make a 70/30 train test split using the train_test_split function imported earlier.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# I import the LinearRegression class from Sklearn
from sklearn.linear_model import LinearRegression
```

```python
# I create the model by storing it in the model variable
model = LinearRegression()
```

```python
# I fit the model to the train set
model.fit(X_train, y_train)
```

```python
# I store the predictions of the X_test split in the y_pred 
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.intercept_)
print(model.coef_)
```

$y_{predicted} = intercept + \beta_0x_1 + \beta_1x_2 + ... + \beta_kx_k$ where $\beta_i$ are the coefficients. 

#### The formula for the linear regression formula for the model is:

y_predicted = -18.7143 + 0.3243x_1 + 0.000002340x_2 + ... + 1.1331x_3 

$\beta_1$ are SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT

$\beta_2$ are ENROLL_8

$\beta_3$ are AVG_READING_8_SCORE

This suggests that reading scores are strongly associated to higher and the amount of enrollees has very limited positive impact on the math score. The correlation between enrollees and math score is also clear from the extra visualisation below.

```python
# Deciding the size of the figure
plt.figure(figsize=(10, 6))


# Plot the data with labels for the legend
plt.scatter(x=df["ENROLL_8"], y=df["AVG_MATH_8_SCORE"], alpha=0.5)

# Labeling the viz
plt.xlabel("Enrollees in 8th Grade")
plt.ylabel("8th Grade Math Score")
plt.title("Scatter Plot of 8th Enrolees vs. 8th Grade Math Score")
plt.grid(True)

# Showing the viz
plt.show()
```

```python
# I use the score method to get the R^2 score. 
# The result shows that model explains 79.9 of the variability in the target variable 
model.score(X_test, y_test)
```

```python
# Mean error which shows the model on average overestimates the prediction
np.mean(model.predict(X_test)-y_test)
```

```python
# Mean absolute error which shows the model is on average off by 2.77
np.mean(np.abs(model.predict(X_test)-y_test))
```

```python
# Root Mean Squared Error (RMSE) -- Penalizes large errors
np.mean((model.predict(X_test)-y_test)**2)**0.5
```
